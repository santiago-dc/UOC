### 2.1 Breast Cancer Wisconsin Data set

El dataset elegido es una ver sión reducida de uno de los más conocidos en la comunidad. Es un dataset enfocado a la clasificación de tumores como malignos o benignos. Mi elección radica en que creo que este dataset es apto para todos los modelos de la PRAC2. Para el caso de la generación de reglas he escogido otro dataset. En este dataset probablemente se implementarán los siguientes modelos:

1. Kmeans. El cual está basado en distancia euclídea. Al tener las etiquetas en el dataset, será sencillo comprobar su eficacia.

2. DBSCAN. Este método de clustering se basa en la desidad de los datos, y no en la distancia entre ellos. También es no supervisado.

3. NN. Para el modelo supervisado implementaré una red neuronal. Estoy convencido de que será el modelo que mejor precisión muestre. 


#### Carga y primer vistazo
Las columnas del dataset corresponden a:
- PatientID
- Clump Thickness
- Uniformity of Cell Size
- Uniformity of Cell Shape
- Marginal Adhesion
- Single Epithelial Cell Size
- Bare Nuclei
- Bland Chromatin
- Normal Nucleoli
- Mitoses

```python
import pandas as pd
df = pd.read_csv('wisconsin-cancer-dataset.csv',header=None,names=['PatientID','Clump Thickness','Uniformity of Cell Size','Uniformity of Cell Shape','Marginal Adhesion','Single Epithelial Cell Size','Bare Nuclei','Bland Chromatin','Normal Nucleoli','Mitoses','Output'])
df.head(10)
```

La columna de salida tiene los valores de 2 y 4 en lugar de 0 y 1. Es necesario remover la primera columna, el id del paciente no aporta información. 

```python
df.describe()
```

Vemos que todos los datos se comprenden entre 1 y 10. Igualmente regularizaremos tras la discretización.

```python
df.dtypes
```

Vemos que hay algo raro en Bare Nuclei, no es de tipo numérico.

```python
df['Bare Nuclei'].value_counts()
```

Removemos las 16 filas con valor nulo.

```python
df = df[df['Bare Nuclei'] != '?']
df=df.astype('int64')
df.dtypes
```

#### Transformación

```python
#replazo el 2 por 0 y el 4 por 1 de la columna 10
df.iloc[:,10].replace(2, 0,inplace=True)
df.iloc[:,10].replace(4, 1,inplace=True)
#eliminamos las filas con valores vacíos o nulos
df.dropna(inplace=True)
#eliminamos la primera columna
df = df.drop('PatientID', 1)
df.head(10)
```

#### Análisis 
Con el objetivo de discretizar alguna variable si es necesario, vamos a analizar los datos en comparación con la relación que tienen sobre la columna 10. Recordemos que dicha columna es en la que se indica si el tumor es maligno o benigno. Para hacer el notebook menos largo solo se muestran los atributos que se pueden discretizar.

```python
df_plots = df.astype({'Output':'str'}) 
from plotnine import *
#Cell Shape
ggplot(df_plots) + geom_bar(aes(x='Uniformity of Cell Shape', fill='Output'),position="fill")
```

Los valores del 5 al 10 se pueden comprimir en 3. Dejando como resultado valores entre el 0 y 7.

```python
df['Uniformity of Cell Shape']=df['Uniformity of Cell Shape'].apply(lambda x: 5 if x==6 else(6 if x==8 or x==7 else(7 if x>8 else x)))
df['Uniformity of Cell Shape'].head(5)
```

```python
#Marginal Adhesion
ggplot(df_plots) + geom_bar(aes(x='Marginal Adhesion', fill='Output'),position="fill")
```

Los valores 4,5,6 y 7,8 también presentan mucha similitud en relación con la salida.Reduciremos estos 4 valores a 2.

```python
df['Marginal Adhesion']=df['Marginal Adhesion'].apply(lambda x: 4 if x>4 and x<7 else(5 if x==8 or x==7 else(6 if x==9 else(7 if x==10 else x))))
df['Marginal Adhesion'].head(5)
```

```python
#Bland Chromatin
ggplot(df_plots) + geom_bar(aes(x='Bland Chromatin', fill='Output'),position="fill")
```

En la columna Bland Chromatin, reducimos los 6 ultimos valores a 2.

```python
df['Bland Chromatin']=df['Bland Chromatin'].apply(lambda x: 5 if x>4 and x<8 else(6 if x>7 else x))
df['Bland Chromatin'].head(5)
```

#### Normalización (min-max)

```python
column_maxes = df.iloc[:,0:9].max()
df_max = column_maxes.max()
normalized_df = df.iloc[:,0:9] / df_max
normalized_df['Output']=df['Output']
normalized_df
```

#### Nueva variable
Resulta evidente mirando las gráficas, que por lo general a mayor valor en las variables, más probabilidades de que la salida sea igual a 1.

```python
col_list= list(normalized_df)
col_list.remove('Output')
normalized_df['Suma'] = normalized_df[col_list].sum(axis=1)
normalized_df
```

```python
df_plots = normalized_df.astype({'Output':'str'}) 
ggplot(df_plots, aes(x='Suma', color='Output', fill='Output')) + geom_density(alpha=0.1)
```

```python
#Volvemos a normalizar y cambiamos el orden de las columnas, dejando output la última.
normalized_df=normalized_df[['Clump Thickness','Uniformity of Cell Size','Uniformity of Cell Shape','Marginal Adhesion','Single Epithelial Cell Size','Bare Nuclei','Bland Chromatin','Normal Nucleoli','Mitoses','Suma','Output']]
column_maxes = normalized_df.iloc[:,0:10].max()
df_max = column_maxes.max()
normalized_df = normalized_df.iloc[:,0:10] / df_max
normalized_df['Output']=df['Output']
normalized_df
```

#### PCA
De cara a aplicar PCA previo a algún modelo (aún por decidir), prodecemos a realizar el análisis. Nótese que este paso puede ser de gran ayuda para visualizar los clusters.

Comenzamos estandarizando los datos. Media ~= 0 y varianza = 1.

```python
from sklearn.preprocessing import StandardScaler
x_std = StandardScaler().fit_transform(normalized_df.iloc[:,0:10])
```

Creamos una instanciación PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
principalComponents = pca.fit_transform(x_std)
```

Mostramos las varianzas de los componentes del PCA.

```python
from matplotlib import pyplot as plt
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_ratio_, color='black')
plt.xlabel('PCA features')
plt.ylabel('variance %')
plt.xticks(features)
plt.show()
```

Lo guardamos.

```python
PCA_components = pd.DataFrame(principalComponents)
```

Parece que el primer componente del PCA explica la mayoría de variación en los datos. En la práctica utilizaremos los dos primeros para mostrar más del 70% de la información total con solo dos componentes.

```python
plt.scatter(PCA_components[0], PCA_components[1],alpha=.1, color='blue')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
```

Esta imagen nos muestra que la información de dataset puede ser claramente agrupada.

#### Transformación a transaction
Ahora procedemos a transformar los datos de cara a utilizar le modelo para generar las reglas de asociación.

Primero creamos una matriz con los objetos de cada compra

```python
import numpy as np
transactions = []
for i in range(0, 7501):
    transactions.append([str(df2.values[i,j]) for j in range(0, 20)])
transactions = np.array(transactions)
transactions.shape
```

```python
transactions[0:5]
```

Utilizando el transaction encoder, se crea una columna para cada alimento, en cada compra se le asigna verdadero o falso a cada registro en función de si está o no en la compra.


```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
data = te.fit_transform(transactions)
data = pd.DataFrame(data, columns = te.columns_)

# getting the shape of the data
data.shape
```

```python
data.head()
```
